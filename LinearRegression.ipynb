```python
//82%
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
import pandas as pd
import numpy as np
```

```python
cols = [f'C{i}' for i in range(1,57) ]
int_cols = [f'C{i}' for i in range(1,12) ]
```

```python
df_train = pd.read_csv('train.csv',names=cols)
df_test = pd.read_csv('test.csv',names=cols[:55])
```

```python
sc = StandardScaler()
df_train[int_cols] = sc.fit_transform(df_train[int_cols])
df_test[int_cols] = sc.fit_transform(df_test[int_cols])
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
model_lr = LinearRegression()
```

```python
model_lr.fit(df_train.drop('C56',axis=1),df_train['C56'])
```

```python
y_pred = model_lr.predict(df_test)
```

```python
data_dict = {'key':df_test['C1'],'score':y_pred}
pd.DataFrame().from_dict(data_dict).to_csv('output_lr.csv',index=False)
```
